##1. Numpy Library

```python
#Declaration of Numpy

import numpy as np

a = np.array([10,20,30,40,50])
b = np.array([[1,2,3],[4,5,6]])

print(a)
print(b)
```

```python
#Dimension of Numpy

print(a.ndim)
print(b.ndim)
```

```python
#Shape of Numpy

print(a.shape)
print(b.shape)
```

```python
#Data type of Numpy

print(a.dtype)
print(b.dtype)
```

###In-built functions using Numpy

```python
#Zero matrix

np.zeros((5,7))
```

```python
#Unity Matrix

np.ones((4,4))
```

```python
np.eye((5))
```

```python
np.arange(1,20,2)
```

```python
np.linspace(0,1,9)
```

```python
c = a*2
d = a+5
print(c)
print(d)
```

```python
#Finding mean value

print(np.mean(a))
print(np.mean(b))
```

```python
#Finding median value

print(np.median(a))
print(np.median(b))
```

```python
#Finding max value

print(np.max(a))
print(np.max(b))
```

```python
#Finding minimum value

print(np.min(a))
print(np.min(b))
```

```python
print(np.sum(a))
print(np.sum(b))
```
