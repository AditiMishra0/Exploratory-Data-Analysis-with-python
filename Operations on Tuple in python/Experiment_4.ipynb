```python
#Python Tuples
#Access Tuples Items

thistuple = ("Cricket", "Football", "Basketball","Volleyball")
print(thistuple[1])
```

```python
#Tuple doesnt allow changes

Sports = ("Cricket", "Football", "Basketball","Volleyball")
Sports[1]= "Tennis"
print(Sports)
```

```python
#Addition of tupples
t1 = (8,9,10)
t2 = (4,5,6)
t3 = t1+t2
print(t3)
```

```python
#Add tuple to a tuple
Sports = ("Cricket", "Football", "Basketball","Volleyball")
y = ("Tennis",)
Sports += y

print(Sports)
```

```python
#Address verification for tuple
Sports = ("Cricket", "Football", "Basketball","Volleyball")
print("Before:",id(Sports))
y= ("Tennis",)
Sports += y
print("After:",id(Sports))
```

```python
#Address verification for lists
Sports = ["Cricket", "Football", "Basketball","Volleyball"]
print("Before:",id(Sports))
y= ["Tennis"]
Sports += y
print("After:",id(Sports))
```
