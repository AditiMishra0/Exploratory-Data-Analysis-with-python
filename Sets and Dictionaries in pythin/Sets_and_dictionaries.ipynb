##Set and its operations in python

```python
#Creation of set
thisset = {"Tennis","Cricket","Football"}
print(thisset)
```

```python
#Duplicates not allowed
thisset = {"Tennis","Cricket","Football","Tennis"}
print(thisset)
```

```python
#The values True and 1 are considered the same value in sets
thisset = {"Tennis","Cricket","Football",True,1,2}
print(thisset)
```

```python
#A set with strings, integers and boolean values
setA = {"Rahul", 23, True, 567, "Female"}
setA
```

```python
#To check membership
thisset = {"Tennis","Cricket","Football", "Pool", "Swimming","Golf"}
print("Tennis" in thisset)
print("Pickleball" in thisset)
```

```python
#Once a set is created, you can't replace its items but can add new items
thisset = {"Tennis","Cricket","Football", "Pool", "Swimming","Golf"}
thisset.add("Pickleball")
print(thisset)
```

```python
#Remove items
thisset = {"Tennis","Cricket","Football", "Pool", "Swimming","Golf"}
thisset.remove("Swimming")
print(thisset)
```

```python
#Set operations
A = {3,4,5,6,7,8}
B = {5,6,7,8,9,10}
print("Union:", A|B)
print("Intersection", A&B)
print("Difference", A-B)
print("Symmetric Difference", A^B)
```

```python
#Frozen set
X = frozenset({"Monday","Tuesday","Wednesday","Thursday","Friday"})
print(X)
print(type(X))
```

```python
#Addition and removal of elements are not allowed in Frozenset
X = frozenset({"Monday","Tuesday","Wednesday","Thursday","Friday"})
X.add("Saturday")
X.remove("Monday")
```

##Dictionaries and its operations in python

```python
#Creating a dictionary
thisdict = {
  "brand": "Apple",
  "model": "Iphone 16",
  "year": 2024
}
thisdict
```

```python
#Retriving value from dictionary
thisdict = {
  "brand": "Apple",
  "model": "Iphone 16",
  "year": 2024
}
x = thisdict["model"]
print("Model of the phone is",x)
```

```python
#Dictionary takes the lastest value for the key (It does not allow duplication)
thisdict = {
  "brand": "Apple",
  "model": "Iphone 16",
  "year": 2024,
  "year": 2025
}
print(thisdict)
```

```python
#Length of key value pairs
print ("Number of key value pairs in the given dictionary is -",len(thisdict))
```

```python
#Changing/Adding key value in dictionary
phone = {
  "brand": "Apple",
  "model": "Iphone",
  "year": 2024
}
phone["colour"] = "Ultramarine",
x = phone.values()
print(x) #after change (only printing values)
```

```python
#Retrival of values
PHONE = {
  "brand": "Apple",
  "model": "Iphone",
  "year": 2024
}
print("model", PHONE["model"])
```

```python
#Problem statement 2 - A dictionary stores student names and marks. Write a python program to search for a student's marks using the name entered , If the student is not found - display a suitable message
students = {
   "A" : 90,
   "B" : 87,
   "C" : 76,
   "D" : 65,
   "E" : 54,
   "F" : 78,
}
name = input("Enter the name of the student - ")
if name in students:
  print("The marks of",name,"are",students[name])
else:
  print ("Student name not found")
```

```python
#Problem statement 2 - using get function
students = {
   "A" : 90,
   "B" : 87,
   "C" : 76,
   "D" : 65,
   "E" : 54,
   "F" : 78,
}
name = input("Enter the name of the student - ")
print(students.get(name, "Student name not found"))
```

```python
#Problem statement 4 - A dictionary stores student names and marks. Write a python program to find the student who scored the highest marks,
StudentMarks = {
    "Aditi" : 98,
    "Arya" : 7,
    "Tanvi": 8,
    "Sankalp" : 9,
}
highestmarks = max(StudentMarks, key=StudentMarks.get)
print("Student with the highest marks - ", highestmarks)
print("The highest marks are:", StudentMarks[highestmarks])
```
