```python
##Python lists
thislist = ["north", "south", "east","west"]
print(thislist)
```

```python
#A list with strings, integers and boolean values
list = ["north", "south", 1,3,5, True, False]
```

```python
#Acess Items
list = ["north", "south", 1,3,5, True, False]
print(list[5])
print(list[1])
```

```python
#Reverse/Negetive indexing
list = ["north", "south", 1,3,5, True, False]
print(list[-1])
print(list[-7])
```

```python
#Sliciing
vegetables = ["carrots", "raddish", "onion", "potatoes", "spinach", "beetroot","cabbage"]
print(vegetables[1:5])
print(vegetables[:4])
print(vegetables[-3:])
```

```python
#Change Item Value
vegetables = ["carrots", "raddish", "onion", "potatoes", "spinach", "beetroot","cabbage"]
vegetables[4] = "cauliflower"
vegetables[0]= "brinjal"
print(vegetables)
```

```python
#Append Items
vegetables = ["carrots", "raddish", "onion", "potatoes", "spinach"]
vegetables.append("drumstick")
print(vegetables)
```

```python
#Insert Items
vegetables = ["carrots", "raddish", "onion", "potatoes", "spinach"]
vegetables.insert(3, "brinjal")
print(vegetables)
```

```python
#Extend list
vegetables = ["carrots", "raddish", "onion", "potatoes", "spinach"]
fruits = ["apple", "banana", "cherry"]
vegetables.extend(fruits)
print(vegetables)
```

```python
#Practice problem 1
numbers = [10,20,30,40,50]

print("Elements from index 1 to 3:")
print(numbers[1:4],"\n")

print("First 3 elements:")
print(numbers[ :3],"\n")

print("Elements from index 2 onwads:")
print(numbers[2:],"\n")

print("All elements in the list:")
print(numbers[:] ,"\n")
```

```python
#Practice problem 2

StudentMarks = [80,88,95,98,76,68,54,92,56,87,74,84]

print("Highest marks = ", max(StudentMarks))
print("Lowest marks = ", min(StudentMarks))
print("Average marks = ", sum(StudentMarks)/len(StudentMarks))
StudentMarks.sort()
print("Sorted marks = ",StudentMarks)
```

```python
#Practice problem 3
Grocery = ["Flour","Lentils","Tissues","Fruits","Vegetables","Eggs","Juice","Dog food","Shampoo","Biscuits","Milk"]
extraItems = ["Detergent","Soaps","Toothpaste","Brush"]
Grocery.extend(extraItems)
print(Grocery,"\n")
Grocery.remove("Tissues")
print("Final grocery list:", Grocery,"\n")

```
